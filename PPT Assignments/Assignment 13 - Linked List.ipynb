```python
#Q1
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_new_list(list1, list2):
    ptr1 = list1
    ptr2 = list2
    newList = None
    tail = None

    while ptr1 is not None and ptr2 is not None:
        if ptr1.data >= ptr2.data:
            new_node = Node(ptr1.data)
            ptr1 = ptr1.next
        else:
            new_node = Node(ptr2.data)
            ptr2 = ptr2.next

        if newList is None:
            newList = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    while ptr1 is not None:
        new_node = Node(ptr1.data)
        tail.next = new_node
        tail = new_node
        ptr1 = ptr1.next

    while ptr2 is not None:
        new_node = Node(ptr2.data)
        tail.next = new_node
        tail = new_node
        ptr2 = ptr2.next

    return newList



# Example 1
list1 = Node(5)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(8)

list2 = Node(1)
list2.next = Node(7)
list2.next.next = Node(4)
list2.next.next.next = Node(5)

new_list = create_new_list(list1, list2)
print("New list: ", end="")
while new_list is not None:
    print(new_list.data, end="->")
    new_list = new_list.next

# Example 2
list1 = Node(2)
list1.next = Node(8)
list1.next.next = Node(9)
list1.next.next.next = Node(3)

list2 = Node(5)
list2.next = Node(3)
list2.next.next = Node(6)
list2.next
```

```python
#Q2
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def remove_duplicates(head):
    if head is None or head.next is None:
        return head

    current = head
    while current is not None and current.next is not None:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head



# Example 1
list1 = Node(11)
list1.next = Node(11)
list1.next.next = Node(11)
list1.next.next.next = Node(21)
list1.next.next.next.next = Node(43)
list1.next.next.next.next.next = Node(43)
list1.next.next.next.next.next.next = Node(60)

print("Original list: ", end="")
current = list1
while current is not None:
    print(current.data, end="->")
    current = current.next

new_list = remove_duplicates(list1)

print("\nModified list: ", end="")
current = new_list
while current is not None:
    print(current.data, end="->")
    current = current.next
# Output: 11->21->43->60

# Example 2
list2 = Node(10)
list2.next = Node(12)
list2.next.next = Node(12)
list2.next.next.next = Node(25)
list2.next.next.next.next = Node(25)
list2.next.next.next.next.next = Node(25)
list2.next.next.next.next.next.next = Node(34)

print("\nOriginal list: ", end="")
current = list2
while current is not None:
    print(current.data, end="->")
    current = current.next

new_list = remove_duplicates(list2)

print("\nModified list: ", end="")
current = new_list
while current is not None:
    print(current.data, end="->")
    current = current.next
```

```python
#Q3
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_list(head):
    prev = None
    current = head

    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

def reverse_k_nodes(head, k):
    if head is None or head.next is None:
        return head

    current = head
    count = 0

    while current is not None and count < k:
        current = current.next
        count += 1

    if count < k:
        return reverse_list(head)

    new_head = reverse_list(head)
    head.next = reverse_k_nodes(current, k)

    return new_head



# Example 1
list1 = Node(1)
list1.next = Node(2)
list1.next.next = Node(2)
list1.next.next.next = Node(4)
list1.next.next.next.next = Node(5)
list1.next.next.next.next.next = Node(6)
list1.next.next.next.next.next.next = Node(7)
list1.next.next.next.next.next.next.next = Node(8)

print("Original list: ", end="")
current = list1
while current is not None:
    print(current.data, end="->")
    current = current.next

k = 4
new_head = reverse_k_nodes(list1, k)

print("\nModified list: ", end="")
current = new_head
while current is not None:
    print(current.data, end="->")
    current = current.next


# Example 2
list2 = Node(1)
list2.next = Node(2)
list2.next.next = Node(3)
list2.next.next.next = Node(4)
list2.next.next
```

```python
#Q4
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_list(head):
    prev = None
    current = head

    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

def reverse_alternate_k_nodes(head, k):
    if head is None or head.next is None:
        return head

    current = head
    count = 0

    while current is not None and count < k:
        current = current.next
        count += 1

    if count < k:
        return head

    new_head = reverse_list(head)
    head.next = current

    count = 0
    while count < k - 1 and current is not None:
        current = current.next
        count += 1

    if current is not None:
        current.next = reverse_alternate_k_nodes(current.next, k)

    return new_head



# Example
list1 = Node(1)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(4)
list1.next.next.next.next = Node(5)
list1.next.next.next.next.next = Node(6)
list1.next.next.next.next.next.next = Node(7)
list1.next.next.next.next.next.next.next = Node(8)
list1.next.next.next.next.next.next.next.next = Node(9)

print("Original list: ", end="")
current = list1
while current is not None:
    print(current.data, end="->")
    current = current.next

k = 3
new_head = reverse_alternate_k_nodes(list1, k)

print("\nModified list: ", end="")
current = new_head
while current is not None:
    print(current.data, end="->")
    current = current.next
```

```python
#Q5
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def delete_last_occurrence(head, key):
    if head is None:
        return None

    prev = None
    last = None

    current = head
    while current is not None:
        if current.data == key:
            last = prev
        prev = current
        current = current.next

    if last is None:
        return head

    if last is None:
        head = head.next
    else:
        last.next = last.next.next

    return head


# Example
list1 = Node(1)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(5)
list1.next.next.next.next = Node(2)
list1.next.next.next.next.next = Node(10)

print("Original list: ", end="")
current = list1
while current is not None:
    print(current.data, end="->")
    current = current.next

key = 2
new_head = delete_last_occurrence(list1, key)

print("\nModified list: ", end="")
current = new_head
while current is not None:
    print(current.data, end="->")
    current = current.next
```

```python
#Q6
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def merge_sorted_lists(a, b):
    if a is None:
        return b
    if b is None:
        return a

    dummy_head = Node(None)
    current = dummy_head

    while a is not None and b is not None:
        if a.data <= b.data:
            current.next = a
            a = a.next
        else:
            current.next = b
            b = b.next
        current = current.next

    if a is not None:
        current.next = a
    if b is not None:
        current.next = b

    return dummy_head.next


# Example
list1 = Node(5)
list1.next = Node(10)
list1.next.next = Node(15)

list2 = Node(2)
list2.next = Node(3)
list2.next.next = Node(20)

print("List 1: ", end="")
current = list1
while current is not None:
    print(current.data, end="->")
    current = current.next

print("\nList 2: ", end="")
current = list2
while current is not None:
    print(current.data, end="->")
    current = current.next

merged_head = merge_sorted_lists(list1, list2)

print("\nMerged List: ", end="")
current = merged_head
while current is not None:
    print(current.data, end="->")
    current = current.next

    
```

```python
#Q7
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverse_doubly_linked_list(head):
    if head is None or head.next is None:
        return head

    current = head
    prev = None

    while current is not None:
        next_node = current.next
        current.next = prev
        current.prev = next_node
        prev = current
        current = next_node

    return prev



# Example
list1 = Node(10)
list1.next = Node(8)
list1.next.prev = list1
list1.next.next = Node(4)
list1.next.next.prev = list1.next
list1.next.next.next = Node(2)
list1.next.next.next.prev = list1.next.next

print("Original list: ", end="")
current = list1
while current is not None:
    print(current.data, end=" ")
    current = current.next

new_head = reverse_doubly_linked_list(list1)

print("\nReversed list: ", end="")
current = new_head
while current is not None:
    print(current.data, end=" ")
    current = current.next
```

```python
#Q8
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def delete_node_at_position(head, position):
    if head is None:
        return head

    if position == 1:
        new_head = head.next
        if new_head is not None:
            new_head.prev = None
        return new_head

    current = head
    prev = None
    count = 1

    while current is not None and count < position:
        prev = current
        current = current.next
        count += 1

    if current is None:
        return head

    prev.next = current.next

    if current.next is not None:
        current.next.prev = prev

    return head



# Example
list1 = Node(1)
list1.next = Node(5)
list1.next.prev = list1
list1.next.next = Node(2)
list1.next.next.prev = list1.next
list1.next.next.next = Node(9)
list1.next.next.next.prev = list1.next.next

print("Original list: ", end="")
current = list1
while current is not None:
    print(current.data, end=" ")
    current = current.next

position = 1
new_head = delete_node_at_position(list1, position)

print("\nModified list: ", end="")
current = new_head
while current is not None:
    print(current.data, end=" ")
    current = current.next
```
