```python
#Q1
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

def removeLoop(head):
    if head is None or head.next is None:
        return head

    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break

    if slow != fast:
        return head

    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    fast.next = None

    return head



# Example 1
head1 = ListNode(1)
head1.next = ListNode(3)
head1.next.next = ListNode(4)
head1.next.next.next = head1.next  
removeLoop(head1) 

# Example 2
head2 = ListNode(1)
head2.next = ListNode(8)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(4)
removeLoop(head2)  

# Example 3
head3 = ListNode(1)
head3.next = ListNode(2)
head3.next.next = ListNode(3)
head3.next.next.next = ListNode(4)
head3.next.next.next.next = head3  
removeLoop(head3)  
```

```python
#Q2
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

def reverseLinkedList(head):
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    return prev

def addOne(head):
    if head is None:
        return ListNode(1)

    reversed_head = reverseLinkedList(head)
    curr = reversed_head
    carry = 1

    while curr:
        curr.val += carry
        carry = curr.val // 10
        curr.val %= 10
        if not curr.next:
            break
        curr = curr.next

    if carry:
        curr.next = ListNode(carry)

    return reverseLinkedList(reversed_head)



# Example 1
head1 = ListNode(4)
head1.next = ListNode(5)
head1.next.next = ListNode(6)
addOne(head1)  # Output: 4->5->7

# Example 2
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
addOne(head2)  # Output: 1->2->4
```

```python
#Q3
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.bottom = None

def merge(a, b):
    if a is None:
        return b
    if b is None:
        return a

    result = None
    if a.val <= b.val:
        result = a
        result.bottom = merge(a.bottom, b)
    else:
        result = b
        result.bottom = merge(a, b.bottom)

    return result

def flattenLinkedList(head):
    if head is None or head.next is None:
        return head

    head.next = flattenLinkedList(head.next)
    head = merge(head, head.next)

    return head

def printLinkedList(head):
    curr = head
    while curr:
        print(curr.val, end=" ")
        curr = curr.bottom
    print()

# Example 1
head1 = ListNode(5)
head1.next = ListNode(10)
head1.next.next = ListNode(19)
head1.next.next.next = ListNode(28)

head1.bottom = ListNode(7)
head1.next.bottom = ListNode(20)
head1.next.next.bottom = ListNode(22)
head1.next.next.next.bottom = ListNode(35)

head1.bottom.bottom = ListNode(8)
head1.next.next.bottom.bottom = ListNode(50)

head1.bottom.bottom.bottom = ListNode(30)
head1.next.next.bottom.bottom.bottom = ListNode(45)

flattened_list1 = flattenLinkedList(head1)
printLinkedList(flattened_list1)

# Example 2
head2 = ListNode(5)
head2.next = ListNode(10)
head2.next.next = ListNode(19)
head2.next.next.next = ListNode(28)

head2.bottom = ListNode(7)
head2.next.next.bottom = ListNode(22)

head2.bottom.bottom = ListNode(8)
head2.next.next.bottom.bottom = ListNode(50)

head2.bottom.bottom.bottom = ListNode(30)

flattened_list2 = flattenLinkedList(head2)
printLinkedList(flattened_list2)
```

```python
#Q4
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.random = None

def copyRandomList(head):
    if head is None:
        return None

    curr = head
    while curr:
        new_node = Node(curr.val)
        new_node.next = curr.next
        curr.next = new_node
        curr = curr.next.next

    curr = head
    while curr:
        if curr.random:
            curr.next.random = curr.random.next
        curr = curr.next.next

    new_head = head.next
    curr = head
    while curr.next:
        temp = curr.next
        curr.next = curr.next.next
        curr = temp

    return new_head


# Example 1
head1 = Node(1)
head1.next = Node(2)
head1.next.next = Node(3)
head1.next.next.next = Node(4)

head1.random = head1.next
head1.next.random = head1.next.next.next

copied_list1 = copyRandomList(head1)
print(copied_list1.val) 

# Example 2
head2 = Node(1)
head2.next = Node(3)
head2.next.next = Node(5)
head2.next.next.next = Node(9)

head2.random = head2
head2.next.next.random = head2.next.next.next

copied_list2 = copyRandomList(head2)
print(copied_list2.val) 
```

```python
#Q5
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

def oddEvenList(head):
    if head is None or head.next is None:
        return head

    odd = head
    even = head.next
    even_head = even  

    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = even_head 

    return head


head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

reordered_list = oddEvenList(head)
while reordered_list:
    print(reordered_list.val, end=" ") 
    reordered_list = reordered_list.next
```

```python
#Q6
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

def rotateLeft(head, k):
    if head is None or head.next is None or k == 0:
        return head

    length = 0
    curr = head
    while curr:
        length += 1
        if length == k:
            kth_node = curr
        if curr.next is None:
            curr.next = head
            break
        curr = curr.next

    head = kth_node.next
    kth_node.next = None

    return head


# Example 1
head1 = ListNode(2)
head1.next = ListNode(4)
head1.next.next = ListNode(7)
head1.next.next.next = ListNode(8)
head1.next.next.next.next = ListNode(9)

shifted_list1 = rotateLeft(head1, 3)
while shifted_list1:
    print(shifted_list1.val, end=" ") 
    shifted_list1 = shifted_list1.next
print()

# Example 2
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(4)
head2.next.next.next.next = ListNode(5)
head2.next.next.next.next.next = ListNode(6)
head2.next.next.next.next.next.next = ListNode(7)
head2.next.next.next.next.next.next.next = ListNode(8)

shifted_list2 = rotateLeft(head2, 4)
while shifted_list2:
    print(shifted_list2.val, end=" ")  
    shifted_list2 = shifted_list2.next
```

```python
#Q7
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

def nextLargerNodes(head):
    values = []
    while head:
        values.append(head.val)
        head = head.next

    n = len(values)
    result = [0] * n
    stack = []

    for i in range(n-1, -1, -1):
        while stack and values[stack[-1]] <= values[i]:
            stack.pop()

        if stack:
            result[i] = values[stack[-1]]

        stack.append(i)

    return result[::-1]


# Example 1
head1 = ListNode(2)
head1.next = ListNode(1)
head1.next.next = ListNode(5)

result1 = nextLargerNodes(head1)
print(result1)  

# Example 2
head2 = ListNode(2)
head2.next = ListNode(7)
head2.next.next = ListNode(4)
head2.next.next.next = ListNode(3)
head2.next.next.next.next = ListNode(5)

result2 = nextLargerNodes(head2)
print(result2)  
```

```python
#Q8
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

def removeZeroSumSublists(head):
    dummy = ListNode(0)
    dummy.next = head

    prefix_sum = 0
    prefix_sum_map = {0: dummy}

    while head:
        prefix_sum += head.val

        if prefix_sum in prefix_sum_map:
            prev = prefix_sum_map[prefix_sum]
            prev.next = head.next
            return removeZeroSumSublists(dummy.next)

        prefix_sum_map[prefix_sum] = head
        head = head.next

    return dummy.next



# Example 1
head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(-3)
head1.next.next.next = ListNode(3)
head1.next.next.next.next = ListNode(1)

result1 = removeZeroSumSublists(head1)
while result1:
    print(result1.val, end=" ") 
    result1 = result1.next
print()

# Example 2
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(-3)
head2.next.next.next.next = ListNode(4)

result2 = removeZeroSumSublists(head2)
while result2:
    print(result2.val, end=" ")
    result2 = result2.next
print()

# Example 3
head3 = ListNode(1)
head3.next = ListNode(2)
head3.next.next = ListNode(3)
head3.next.next.next = ListNode(-3)
head3.next.next.next.next = ListNode(-2)

result3 = removeZeroSumSublists(head3)
while result3:
    print(result3.val, end=" ") 
    result3 = result3.next
print()
```
