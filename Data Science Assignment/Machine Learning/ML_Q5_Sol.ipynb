```python
import pandas as pd
df = pd.read_csv("rideshare_kaggle.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=['timezone', 'datetime', 'product_id', 'latitude', 'longitude', 'id'])
df.shape
```

```python
df.describe()
```

```python
df_categorical = df[['source', 'destination', 'cab_type','name','short_summary','long_summary']]
df_categorical.describe()
```

```python
duplics = df.duplicated()
print(df[duplics])
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
cat_f = ['source', 'destination', 'cab_type','name','short_summary','long_summary','icon']
from sklearn.preprocessing import LabelEncoder

for i in cat_f:
    encoder = LabelEncoder()
    df[i] = encoder.fit_transform(df[i])
```

```python
df.info()
```

```python
X = df.drop(columns = "price")
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr.fit(X_train,y_train)
lr_pred = lr.predict(X_test)
mse = mean_squared_error(y_test, lr_pred)
print(mse)
```
